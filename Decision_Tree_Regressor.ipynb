```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
 df=load_diabetes()
df
```

```python
X=pd.DataFrame(df['data'],columns=df['feature_names'])
y=df['target']
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
treeRegressor=DecisionTreeRegressor()
treeRegressor.fit(X_train,y_train)
```

```python
y_pred=treeRegressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
params=dict(criterion=['squared_error','friedman_mse','absolute_error'],splitter=['best','random'],max_depth=[2,3,4,5,10,15,20,25],max_features=['auto','sqrt','log2'])
params
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=params,scoring='neg_mean_squared_error',cv=5)
```

```python
import warnings
```

```python
warnings.filterwarnings("ignore")
grid.fit(X_train,y_train)
y_pred1=grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
print(r2_score(y_test,y_pred1))
```
