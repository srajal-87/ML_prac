```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(n_samples=500,noise=0.05)
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scale=scaler.fit_transform(X)
```

```python
X_scale
```

```python
plt.scatter(X_scale[:,0],X_scale[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster=DBSCAN(min_samples=5,eps=0.10)
cluster.fit_predict(X_scale)
```

```python
cluster.labels_
```

```python
plt.scatter(X_scale[:,0],X_scale[:,1],c=cluster.labels_)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
outlier=LocalOutlierFactor(n_neighbors=10)
outlier.fit_predict(X_scale)
```

```python
outlier.negative_outlier_factor_
```
