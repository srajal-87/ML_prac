# MATPLOTLIB
#### Matplotlib graphs your data on Figures (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more Axes, an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.). The simplest way of creating a Figure with an Axes is using pyplot.subplots. We can then use Axes.plot to draw some data on the Axes, and show to display the figure:
### key components:
Figure: The entire window or page that everything is drawn on. It can contain multiple plots.
Axes: The area on which data is plotted. A figure can contain multiple Axes objects.
Axis: Represents a single x-axis or y-axis in a plot.

#### AXES: 
#### An Axes object in Matplotlib represents a single plot or graph. It contains various elements such as title, labels, ticks, and the graphical representation of data (e.g., lines, bars, scatter points).
##### Creation: When you create a plot using plt.plot() or similar functions, an Axes object is automatically created.
##### Multiple Axes: A single figure can contain multiple Axes objects (e.g., in a subplot layout).


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
## example
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
plt.show()                           # Show the figure.
```

```python
fig = plt.figure()             # an empty figure with no Axes
fig, ax = plt.subplots()       # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

## Coding styles
The explicit and the implicit interfaces
As noted above, there are essentially two ways to use Matplotlib:

Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style").

Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting.

### So one can use the OO-style

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# layout=constrained:the subplot fit in to the figure witout overlapping 

ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```

### or the pyplot-style:

```python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### Making a helper functions
If you need to make the same plots over and over again with different data sets, or want to easily wrap Matplotlib methods, use the recommended signature function below.



```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

## Matplot Plots

## 1 Line Plot
#### Why: Used to visualize the trend of data over continuous intervals. Useful for showing how a variable changes over time or #### space.
#### Attributes:
#### color: Specifies the color of the line.
#### linestyle: Defines the style of the line (e.g., solid, dashed).
#### linewidth: Sets the width of the line.
#### marker: Specifies the marker style for data points.
#### label: Adds a label to the line for the legend.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.linspace(0,5,100)

plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', label='line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

## 2 Scatter Plot
#### Why: Used to visualize the relationship between two numerical variables. Helps in identifying patterns, clusters, or outliers in data.
Attributes:
#### color: Specifies the color of the markers.
#### marker: Defines the shape of the markers (e.g., circles, squares).
#### s: Sets the size of the markers.
#### alpha: Sets the transparency level of the markers.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='red', marker='o', s=50, alpha=0.5, label='Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

## 3 Bar Plot
#### Why: Used to compare categorical data or show relative sizes across categories.
#### Attributes:
#### color: Specifies the color of the bars.
#### width: Sets the width of the bars.
#### edgecolor: Defines the color of the bar edges.
#### align: Specifies the alignment of the bars (e.g., 'center', 'edge').

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values, color='green', width=0.5, edgecolor='black', align='center')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

### 4 Histogram
#### Why: Used to visualize the distribution of numerical data. Shows frequency or probability density of values.
#### Attributes:
#### bins: Specifies the number of bins or intervals.
#### color: Sets the color of the bars.
#### edgecolor: Defines the color of the bar edges.
#### alpha: Sets the transparency level of the bars.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

## 5Pie Chart
#### Why: Used to show the proportion of parts to a whole. Useful for visualizing categorical data distribution.
#### Attributes:
#### labels: Specifies the labels for each wedge.
#### colors: Sets the colors of the wedges.
#### autopct: Formats the percentage labels.
#### startangle: Rotates the start of the pie chart.

```python
import matplotlib.pyplot as plt

sizes = [30, 20, 25, 15]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'],
        autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

# SEABORN
### Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It is built on top of Matplotlib and integrates well with pandas data structures.

## 1. Count Plot
#### Why: Used to show the counts of observations in each categorical bin using bars.
#### Attributes:
#### x, y: Categorical variables for the count distribution.
#### hue: Grouping variable that will produce points with different colors.
#### palette: Color palette to use for plotting.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C', 'A', 'B', 'C', 'C']
})

# Count plot
sns.countplot(x='Category', data=data, hue=None, palette='viridis')
plt.title('Count Plot')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

## 2 Scatter Plot
#### Why: Used to visualize the relationship between two numerical variables.
#### Attributes:
#### x, y: Variables that specify positions on the x and y axes.
#### hue: Grouping variable that will produce points with different colors.
#### palette: Color palette to use for plotting.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 15, 7, 8, 12],
    'Category': ['A', 'B', 'A', 'B', 'A']
})

# Scatter plot
sns.scatterplot(x='X', y='Y', data=data, hue='Category', palette='Set2')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 3 Bar Plot
#### Why: Used to compare categorical data or show relative sizes across categories.
#### Attributes:
#### x, y: Variables that specify positions on the x and y axes.
#### hue: Grouping variable that will produce bars with different colors.
#### palette: Color palette to use for plotting.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [10, 20, 15]
})

# Bar plot
sns.barplot(x='Category', y='Value', data=data, hue=None, palette='pastel')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

## 4Histogram
#### Why: Used to visualize the distribution of numerical data.
#### Attributes:
#### x: Variable that specifies positions on the x-axis.
#### hue: Grouping variable that will produce histograms with different colors.
#### palette: Color palette to use for plotting.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)

# Histogram
sns.histplot(data, bins=30, kde=True, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 5Box Plot
#### Why: Used to visualize the distribution of numerical data across several categories.
#### Attributes:
#### x, y: Variables that specify positions on the x and y axes.
#### hue: Grouping variable that will produce box plots with different colors.
#### palette: Color palette to use for plotting

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C'],
    'Value': [10, 15, 20, 25, 30]
})

# Box plot
sns.boxplot(x='Category', y='Value', data=data, hue=None, palette='colorblind')
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```
