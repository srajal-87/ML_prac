```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
cancer_dataset.feature_names
```

```python
cancer_dataset.target_names
```

```python
np.array(np.unique(cancer_dataset.target))
```

```python
df=pd.DataFrame(data=cancer_dataset['data'],columns=cancer_dataset['feature_names'])
```

```python
df
```

```python
df.iloc[:,0]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df.iloc[:,0],y=df.iloc[:,1],c=cancer_dataset['target'],cmap='plasma')
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
scaled_dataset=scaler.transform(df)
```

```python
scaled_dataset
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=scaled_dataset[:,0],y=scaled_dataset[:,1],c=cancer_dataset['target'],cmap='plasma')
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca.fit(scaled_dataset)
```

```python
pca_data=pca.transform(scaled_dataset)
```

```python
pca.explained_variance_
```

```python
pca_data
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=pca_data[:,0],y=pca_data[:,1],c=cancer_dataset['target'],cmap='plasma')
```
