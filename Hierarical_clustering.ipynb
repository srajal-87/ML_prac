```python
iris_data=load_iris()
iris_data
```

```python
df=pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scale=scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_data=pca.fit_transform(X_scale)
```

```python
plt.scatter(pca_data[:,0],pca_data[:,1],c=iris_data.target)
```

```python
## find the optimal number of clusters using dendrogram
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,8))
sc.dendrogram(sc.linkage(X_scale,method='ward'))
plt.title('Dendrogram plot')
plt.xlabel('customers')
plt.ylabel('Eucleidian distance')
```

```python
## by seen dendrogram plot if we pass horizonatal line through the maximum vectical distance line than it pasees through two points 
## that means the dataset is divided in to two clusters
```

```python
from sklearn.cluster import AgglomerativeClustering
agg=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
agg.fit(pca_data)
```
